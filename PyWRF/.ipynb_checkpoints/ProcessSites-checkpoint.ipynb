```python
import numpy as np
import pandas as pd
#pd.set_option('display.width', 600)           # or a larger number to fit your screen
#pd.set_option('display.max_columns', None)    # ensure no limit on number of columns
#pd.set_option('display.max_rows', None)    # ensure no limit on number of columns
#NOAA station directory
input_file = "/home/rsaltos/GHCNh_Data/GHCNh_USI0000KTCY_2020.psv"
df = pd.read_csv(input_file, delimiter='|',dtype="str", usecols=["Station_name","DATE","LATITUDE","LONGITUDE"
                                                                 ,"temperature","altimeter","dew_point_temperature",
                                                                 "wind_direction","wind_speed","relative_humidity"])
#print(df.head(2))
df['DATE'] = pd.to_datetime(df['DATE'])
# Filter for dates between June 1 and June 30, 2020
start_date = '2020-05-31 23:55:00'
end_date = '2020-07-01'
mask_date = (df['DATE'] >= start_date) & (df['DATE'] <= end_date)

#print(df.head(2))



# Filter for time ending in :55:00
mask_time = df['DATE'].dt.strftime('%M:%S') == '55:00'

# Apply both filters
df = df[mask_date & mask_time]


#print(df.head(2))
print(df.head(2))


# Generate full expected time range (every 1 hour at :55:00 between June 1â€“30)
expected_range = pd.date_range(start='2020-05-31 23:55:00', 
                               end='2020-06-30 23:55:00', 
                               freq='1h')

actual_timestamps = df['DATE']
missing_times = expected_range.difference(actual_timestamps)

# Display missing date-hours
print("Missing date-hours at :55:00:")
print(missing_times)

# Create a DataFrame with missing timestamps and NaNs for all other columns
missing_df = pd.DataFrame({'datetime': missing_times})
# Add NaNs for all other columns (same as df_filtered)
for col in df.columns:
    if col != 'datetime':
        missing_df[col] = pd.NA

# Ensure missing_df has the same columns and order as df
missing_df = pd.DataFrame(columns=df.columns)

# Fill in the 'datetime' column with the missing times
missing_df['DATE'] = missing_times

# Combine with original df
df_complete = pd.concat([df, missing_df], ignore_index=True)

# Sort by datetime (optional)
df_complete = df_complete.sort_values(by='DATE').reset_index(drop=True)



print(df_complete[df_complete['DATE'].isin(pd.to_datetime(
                                ['2020-06-01 11:55:00', '2020-06-01 12:55:00', '2020-06-01 15:55:00', 
                                '2020-06-04 13:55:00', '2020-06-04 14:55:00', '2020-06-06 01:55:00', 
                                '2020-06-22 08:55:00', '2020-06-22 09:55:00']))])

df_complete['DATE'] = df_complete['DATE'].dt.round('h')


print(df_complete[df_complete['DATE'].isin(pd.to_datetime(
                                ['2020-06-01 12:00:00', '2020-06-01 13:00:00', '2020-06-01 16:00:00', 
                                '2020-06-04 14:00:00', '2020-06-04 15:00:00', '2020-06-06 02:00:00', 
                                '2020-06-22 09:00:00', '2020-06-22 10:00:00']))])
df_complete[0:718].to_csv("output.csv", index=False)
```

```python
import numpy as np
import pandas as pd
import pytz
#pd.set_option('display.max_rows', None)    # ensure no limit on number of columns
#NOAA station directory
input_file = "/home/mpasillaspablo/CIMIS/6_Davis_hly2020.csv"
#df = pd.read_csv(input_file, delimiter=',',dtype="str")#,usecols=["Stn Name","CIMIS Region","Date","Hour (PST)"])

pd.set_option('display.width', 600)           # or a larger number to fit your screen
pd.set_option('display.max_columns', None)    # ensure no limit on number of columns
pd.set_option('display.max_rows', None)    # ensure no limit on number of columns

df = pd.read_csv(input_file, delimiter=',',dtype="str", usecols=[
    "Stn Name","Date","Hour (PST)","ETo (mm)","Precip (mm)","Sol Rad (W/sq.m)","Vap Pres (kPa)",
    "Air Temp (C)","Rel Hum (%)","Dew Point (C)","Wind Speed (m/s)","Wind Dir (0-360)","Soil Temp (C)"])

# Step 1: Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
# Step 2: Identify '2400' rows
mask_2400 = df['Hour (PST)'] == '2400'
# Step 3: Replace '2400' with '0000'
df.loc[mask_2400, 'Hour (PST)'] = '0000'
# Step 4: Add 1 day to the Date for those rows
df.loc[mask_2400, 'Date'] = df.loc[mask_2400, 'Date'] + pd.Timedelta(days=1)

# Combine date and hour into one string
datetime_str = df['Date'].dt.strftime('%m/%d/%Y') + ' ' + df['Hour (PST)']

#Parse to datetime (assume PST)
df['Datetime_PST'] = pd.to_datetime(datetime_str, format='%m/%d/%Y %H%M', errors='raise')

pst_fixed = pytz.FixedOffset(-8 * 60)  # -8 hours = -480 minutes
df['Datetime_PST'] = df['Datetime_PST'].dt.tz_localize(pst_fixed)
# Convert to UTC
df.insert(2, 'Datetime', df['Datetime_PST'].dt.tz_convert('UTC'))

start_datetime = '2020-06-01 00:00:00'
end_datetime = '2020-06-30 21:00:00'
mask_datetime = (df['Datetime'] >= start_datetime) & (df['Datetime'] <= end_datetime)


df = df.drop(columns=['Date','Hour (PST)','Datetime_PST'])
print(len(mask_datetime[mask_datetime == True]))
print(df[mask_datetime].tail(2))

Vap_Pres= df['Vap Pres (kPa)'][mask_datetime].astype(float).to_numpy()
print(Vap_Pres)
```

```python
import os
import numpy as np
import pandas as pd
from netCDF4 import Dataset
from wrf import getvar, to_np, ll_to_xy, ALL_TIMES
from glob import glob

#Configurations for all WRF models
wrf_models = [
    ("/data/wto/WRFOUT/Sac_NDown_UCM1_PBL2_FDDA1_250710/", "PBL2_FDDA1", "blue"),
    ("/data/wto/WRFOUT/Sac_NDown_UCM1_PBL5_FDDA1_250711/", "PBL5_FDDA1", "red"),
    ("/data/wto/WRFOUT/Sac_NDown_UCM1_PBL7_FDDA1_250715/", "PBL7_FDDA1", "purple")
]

#NOAA station directory
data_dir = "/home/rsaltos/GHCNh_Data"

#Time range for June 2020
start = pd.Timestamp("2020-06-01 00:00:00")
end = pd.Timestamp("2020-06-30 21:00:00")
full_time_index = pd.date_range(start=start, end=end, freq='h')

#Results: dictionary of station to metrics per model
results = {}

#Loop through each NOAA file
for file in glob(os.path.join(data_dir, "*.psv")):
    try:
        df = pd.read_csv(file, delimiter='|', skiprows=32, usecols=[1,2,3,4,36])
        df.columns = ["Station_name", "DATE", "LATITUDE", "LONGITUDE", "wind_speed"]

        lat = df["LATITUDE"].iloc[0]
        lon = df["LONGITUDE"].iloc[0]
        station = df["Station_name"].iloc[0]

        df["DATE"] = pd.to_datetime(df["DATE"], errors="coerce").dt.round("h")
        df["wind_speed"] = pd.to_numeric(df["wind_speed"], errors="coerce")

        df = df[(df["DATE"] >= start) & (df["DATE"] <= end)].drop_duplicates(subset="DATE")
        df = df.set_index("DATE").reindex(full_time_index)
        df.index.name = "DATE"
        #Convert to m/s
        df["wind_speed"] = df["wind_speed"] * 0.514444
        #Replace NaNs with -999
        df["wind_speed"] = df["wind_speed"].fillna(-9999)

        station_results = {"station": station}

        for model_path, model_label, _ in wrf_models:
            try:
                wrf_files = sorted(glob(os.path.join(model_path, "wrfout_d01_2020-06-*")))
                if not wrf_files:
                    continue

                sample_ds = Dataset(wrf_files[0])
                x, y = ll_to_xy(sample_ds, lat, lon, as_int=True)

                wrf_values = []
                for f in wrf_files:
                    ds = Dataset(f)
                    var = getvar(ds, "wspd10", timeidx=ALL_TIMES)
                    arr = to_np(var)
                    wrf_values.extend(arr[:, y, x] if arr.ndim == 3 else [arr[y, x]])

                wrf_series = pd.Series(wrf_values, index=pd.date_range(start, periods=len(wrf_values), freq='h'))
                merged = df.join(wrf_series.rename("WRF"), how="inner").dropna(subset=["wind_speed", "WRF"])

                ref = merged["wind_speed"].values
                model = merged["WRF"].values
                diff = model - ref

                mean_bias = np.mean(diff)
                rmsd = np.sqrt(np.mean(diff**2))
                denom = np.sum((np.abs(model - np.mean(ref)) + np.abs(ref - np.mean(ref)))**2)
                ioa = 1 - np.sum((model - ref)**2) / denom if denom != 0 else np.nan

                station_results[f"mean_bias_{model_label}"] = mean_bias
                station_results[f"rmsd_{model_label}"] = rmsd
                station_results[f"ioa_{model_label}"] = ioa

            except Exception as e:
                continue  #skip model if it fails

        results[station] = station_results

    except Exception as e:
        continue  #skip station if it fails

#Save to CSV
results_df = pd.DataFrame(results.values())
results_df.to_csv("mean_bias_all_models.csv", index=False)
```
