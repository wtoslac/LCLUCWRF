## Comparing WRF Wind Speed at 10-m with and without Urban Canopy Modeling (UCM) ###
I ran WRF center around Sacramento with Domain 1 with 4km grid size 1000km wide EW and NS. Domain 2 with 1 km grid size 200km by 200km.<br>
The forecasts are made for Jan 1st 2025-00:00:00 to Jan 2nd 2025-00:00:00.<br>
The run without UCM took about 10-minutes to forecast each hour.<br>
The run with UCM took about 35-minutes to forecast each hour. About 3.5X longer. <br>

### In Code box 1 (Domain 2)
This produced the 10-m level Wind Speed in $\frac{m}{s}$ for domain 2 of the forecast.<br>
The figure on the left is with UCM and the right is without.<br>

### In Code box 2 (Domain 1)
This produced the 10-m level Wind Speed in $\frac{m}{s}$ for domain 1 of the forecast.<br>
The figure on the left is with UCM and the right is without.<br>


```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
from wrf import getvar, to_np, get_cartopy, latlon_coords
import cartopy as cp

# Open the NetCDF file
ncfile = Dataset("/data/wto/WRFOUT/run_20250327_Cali_Sacr_2Dom_WithUCM/wrfout_d02_2025-01-02_00:00:00")
ncfileNoUCM = Dataset("/data/wto/WRFOUT/run_20250326_Cali_Sacr_2Dom_NoUCM/wrfout_d02_2025-01-02_00:00:00")

# Extract the U and V components of the wind at the surface (2-m level)
u_component = getvar(ncfile, "U10")  # U component of the wind at 10 meters
v_component = getvar(ncfile, "V10")  # V component of the wind at 10 meters
u_component_no_ucm = getvar(ncfileNoUCM, "U10")
v_component_no_ucm = getvar(ncfileNoUCM, "V10")

# Calculate the wind speed (magnitude of wind vector)
wind_speed = np.sqrt(to_np(u_component)**2 + to_np(v_component)**2)
wind_speed_no_ucm = np.sqrt(to_np(u_component_no_ucm)**2 + to_np(v_component_no_ucm)**2)

# Get projection and coordinates
wrf_proj = get_cartopy(u_component)
lats, lons = latlon_coords(u_component)

# Set global min/max before plotting to keep colors consistent
vmin = np.floor(min(wind_speed.min(), wind_speed_no_ucm.min()))
vmax = np.ceil(max(wind_speed.max(), wind_speed_no_ucm.max()))

# Latitude and Longitude of Sacramento
sacramento_lat = 38.58
sacramento_lon = -121.49

# Create figure with minimal whitespace
fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={'projection': wrf_proj}, 
                               gridspec_kw={'wspace': -0.5}, figsize=(18, 8))

# Reduce extra margins
fig.subplots_adjust(left=0.08, right=0.92, top=0.90, bottom=0.15)

# First contour plot (With UCM)
ax1.set_title("Wind Speed WITH UCM [m/s]")
ax1.set_extent([to_np(lons).min(), to_np(lons).max(), to_np(lats).min(), to_np(lats).max()], crs=cp.crs.PlateCarree())
contour1 = ax1.contourf(to_np(lons), to_np(lats), wind_speed, 20, transform=cp.crs.PlateCarree(), cmap="jet", vmin=vmin, vmax=vmax)
ax1.add_feature(cp.feature.COASTLINE)
ax1.add_feature(cp.feature.STATES)

# Add a star marker for Sacramento
ax1.scatter(sacramento_lon, sacramento_lat, color='black', marker='*', s=100, transform=cp.crs.PlateCarree(), label="Sacramento")

# Second contour plot (Without UCM)
ax2.set_title("Wind Speed WITHOUT UCM [m/s]")
ax2.set_extent([to_np(lons).min(), to_np(lons).max(), to_np(lats).min(), to_np(lats).max()], crs=cp.crs.PlateCarree())
contour2 = ax2.contourf(to_np(lons), to_np(lats), wind_speed_no_ucm, 20, transform=cp.crs.PlateCarree(), cmap="jet", vmin=vmin, vmax=vmax)
ax2.add_feature(cp.feature.COASTLINE)
ax2.add_feature(cp.feature.STATES)

# Add a star marker for Sacramento
ax2.scatter(sacramento_lon, sacramento_lat, color='black', marker='*', s=100, transform=cp.crs.PlateCarree(), label="Sacramento")

# Add shared colorbar spanning only the width of both plots
cbar = fig.colorbar(contour1, ax=[ax1, ax2], orientation="horizontal", fraction=0.085, pad=0.04)
cbar.set_label("Wind Speed [m/s]")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
from wrf import getvar, to_np, get_cartopy, latlon_coords
import cartopy as cp

# Open the NetCDF file
ncfile = Dataset("/data/wto/WRFOUT/run_20250327_Cali_Sacr_2Dom_WithUCM/wrfout_d02_2025-01-02_00:00:00")
ncfileNoUCM = Dataset("/data/wto/WRFOUT/run_20250326_Cali_Sacr_2Dom_NoUCM/wrfout_d02_2025-01-02_00:00:00")

# Extract the 2-m level temperature
VarUCM = getvar(ncfile, "slp")
VarNoUCM = getvar(ncfileNoUCM, "slp")

# Get projection and coordinates
wrf_proj = get_cartopy(VarUCM)
lats, lons = latlon_coords(VarUCM)

# Set global min/max before plotting to keep colors consistent
vmin = np.floor(min(to_np(VarUCM).min(), to_np(VarNoUCM).min()))
#print(vmin)
vmax = np.ceil(max(to_np(VarUCM).max(), to_np(VarNoUCM).max()))
#print(vmax)

# Create figure with minimal whitespace
fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={'projection': wrf_proj}, 
                               gridspec_kw={'wspace': -0.5}, figsize=(18, 8))

# Reduce extra margins
fig.subplots_adjust(left=0.08, right=0.92, top=0.90, bottom=0.15)

# First contour plot (With UCM)
ax1.set_title("Surface Level Pressure WITH UCM [hPa]")
ax1.set_extent([to_np(lons).min(), to_np(lons).max(), to_np(lats).min(), to_np(lats).max()], crs=cp.crs.PlateCarree())
contour1 = ax1.contourf(to_np(lons), to_np(lats), to_np(VarUCM), 20, transform=cp.crs.PlateCarree(), cmap="jet", vmin=vmin, vmax=vmax)
ax1.add_feature(cp.feature.COASTLINE)
ax1.add_feature(cp.feature.STATES)

# Add a star marker for Sacramento
ax1.scatter(-121.49, 38.58, color='black', marker='*', s=100, label="Sacramento",transform=cp.crs.PlateCarree())

# Second contour plot (Without UCM)
ax2.set_title("Surface Level Pressure WITHOUT UCM [hPa]")
ax2.set_extent([to_np(lons).min(), to_np(lons).max(), to_np(lats).min(), to_np(lats).max()], crs=cp.crs.PlateCarree())
contour2 = ax2.contourf(to_np(lons), to_np(lats), to_np(VarNoUCM), 20, transform=cp.crs.PlateCarree(), cmap="jet", vmin=vmin, vmax=vmax)
ax2.add_feature(cp.feature.COASTLINE)
ax2.add_feature(cp.feature.STATES)

# Add a star marker for Sacramento
ax2.scatter(-121.49, 38.58, color='black', marker='*', s=100, label="Sacramento",transform=cp.crs.PlateCarree())

# Add shared colorbar spanning only the width of both plots
cbar = fig.colorbar(contour1, ax=[ax1, ax2], orientation="horizontal", fraction=0.085, pad=0.04)
cbar.set_label("Surface Level Pressure [hPa]")

plt.show()
```
